```python
import time
class Timer:    
    def __enter__(self):
        self.tick = time.time()
        return self

    def __exit__(self, *args, **kwargs):
        self.tock = time.time()
        self.elapsed = self.tock - self.tick

# make an array where each time a model is trained, the name, the time, the accuracy and the confusion matrix are stored
model_results = []
```

#### Creazione del modello AdaboostM1

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier

# Creare un'istanza del classificatore Decision Tree
tree = DecisionTreeClassifier()

class AdaBoostM1:
    def __init__(self, n_estimators=50, learning_rate=1):
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Initialize the weights to 1/n_samples
        w = np.ones(n_samples) / n_samples
        self.estimators_ = []
        self.alphas_ = []

        for i in range(self.n_estimators):
            # Train a weak classifier using the current weights
            estimator = DecisionTreeClassifier(max_depth=1)
            estimator.fit(X, y, sample_weight=w)

            # Calculate the error of the weak classifier
            y_pred = estimator.predict(X)
            error = np.sum(w * (y_pred != y))

            # Calculate the alpha coefficient of the weak classifier
            alpha = self.learning_rate * np.log((1 - error) / error)
            self.alphas_.append(alpha)

            # Update the weights based on the performance of the weak classifier
            w *= np.exp(-alpha * y * y_pred)
            w /= np.sum(w)

            self.estimators_.append(estimator)

    def predict(self, X):
        predictions = np.zeros(len(X))

        for i in range(len(self.estimators_)):
            estimator = self.estimators_[i]
            alpha = self.alphas_[i]
            predictions += alpha * estimator.predict(X)

        return np.sign(predictions)
```

```python
#AdaBoostM1
ada = AdaBoostM1(n_estimators=50, learning_rate=1)
with Timer() as predict_time:
    ada.fit(X_train_np, y_train)
y_pred_adaboost = ada.predict(X_test_np)
accuracy_adaboost = accuracy_score(y_test, y_pred_adaboost)
conf_matrix_adaboost = confusion_matrix(y_test, y_pred_adaboost)
model_results.append(['AdaBoostM1', predict_time.elapsed, accuracy_adaboost, conf_matrix_adaboost])
```

#### Real Ada Boost

- L'algoritmo Real AdaBoost (Real Adaptive Boosting) è una variante dell'algoritmo AdaBoost (Adaptive Boosting) che cerca di correggere alcuni limiti dell'algoritmo originale, in particolare la sensibilità ai valori anomali e la propensione all'overfitting. L'idea principale di Real AdaBoost è di utilizzare la funzione esponenziale di Pérez-Cruz come funzione di costo anziché la funzione di costo binaria utilizzata da AdaBoost.
Di seguito, un esempio di codice per l'algoritmo Real AdaBoost in Python utilizzando la libreria scikit-learn:

- In questo esempio, è utilizzata la classe AdaBoostClassifier della libreria scikit-learn per creare un'istanza del classificatore Real AdaBoost con 100 stimatori (il parametro n_estimators) e utilizzando la versione "real" dell'algoritmo con il parametro algorithm='SAMME.R'. Utilizziamo un albero di decisione con profondità massima 1 come classificatore base (il parametro base_estimator). Addestriamo il classificatore sui dati di addestramento X_train e le loro etichette y_train utilizzando il metodo fit. Successivamente, prediciamo le etichette per i dati di test X_test utilizzando il metodo predict, e valutiamo la precisione del classificatore utilizzando la funzione accuracy_score.
Come per gli altri algoritmi di apprendimento automatico, ci sono molte opzioni di configurazione e iperparametri che possono influire sulle prestazioni del modello. La scelta del classificatore base, del numero di stimatori, della profondità massima degli alberi e di altri parametri può avere un impatto significativo sulle prestazioni del modello. Tali iperparametri possono essere ottimizzati utilizzando la tecnica della ricerca degli iperparametri.

```python
import cupy as cp
from sklearn.tree import DecisionTreeClassifier

class RealAdaBoostGPU:
    def __init__(self, n_estimators=50):
        self.n_estimators = n_estimators
        self.alphas = []
        self.models = []

    def fit(self, X, y):
        # Transfer data to GPU
        X_gpu = cp.asarray(X)
        y_gpu = cp.asarray(y)

        # Initialize weights on GPU
        n_samples = X.shape[0]
        weights = cp.ones(n_samples) / n_samples

        for _ in range(self.n_estimators):
            # Train a weak learner with CPU-based DecisionTreeClassifier
            model = DecisionTreeClassifier(max_depth=1)
            model.fit(cp.asnumpy(X_gpu), cp.asnumpy(y_gpu), sample_weight=cp.asnumpy(weights))

            # Store the model
            self.models.append(model)

            # Predict on training data and move predictions to GPU
            predictions = cp.asarray(model.predict(cp.asnumpy(X_gpu)))

            # Calculate weighted error rate on GPU
            misclassified = (predictions != y_gpu)
            error = cp.sum(weights * misclassified) / cp.sum(weights)

            # Calculate alpha (model weight) using real-valued update on GPU
            alpha = 0.5 * cp.log((1 - error) / error)
            self.alphas.append(alpha)

            # Update weights on GPU
            weights *= cp.exp(-alpha * y_gpu * predictions)
            weights /= cp.sum(weights)  # Normalize weights

    def predict(self, X):
        # Transfer test data to GPU
        X_gpu = cp.asarray(X)
        final_pred = cp.zeros(X.shape[0])

        for alpha, model in zip(self.alphas, self.models):
            # Make predictions on the CPU, then transfer to GPU for accumulation
            predictions = cp.asarray(model.predict(cp.asnumpy(X_gpu)))
            final_pred += alpha * predictions

        # Transfer result back to CPU for final output
        return cp.sign(final_pred).get()
```

```python
# RealAdaBoost GPU
gpu_real_adaboost = RealAdaBoostGPU(n_estimators=50)
with Timer() as predict_time:
    gpu_real_adaboost.fit(X_train_np, y_train)
y_pred_realadaboost_gpu = gpu_real_adaboost.predict(X_test_np)
accuracy_realadaboost_gpu = accuracy_score(y_test, y_pred_realadaboost_gpu)
conf_matrix_realadaboost_gpu = confusion_matrix(y_test, y_pred_realadaboost_gpu)
model_results.append(['RealAdaBoost_GPU', predict_time.elapsed, accuracy_realadaboost_gpu, conf_matrix_realadaboost_gpu])
```

#### Creazione del modello NaiveBayes

```python
import numpy as np

class NaiveBayes:
    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.classes = np.unique(y)
        n_classes = len(self.classes)

        # Inizializza i parametri di media e varianza per ogni classe e attributo
        self.mean = np.zeros((n_classes, n_features))
        self.var = np.zeros((n_classes, n_features))
        self.priors = np.zeros(n_classes)

        # Calcola la media, la varianza e la probabilità a priori per ogni classe e attributo
        for idx, c in enumerate(self.classes):
            X_c = X[y == c]  # Select all rows where the class label is 'c'
            self.mean[idx, :] = X_c.mean(axis=0)
            self.var[idx, :] = X_c.var(axis=0)
            self.priors[idx] = X_c.shape[0] / float(n_samples)

    def predict(self, X):
        y_pred = [self._predict(x) for x in X]
        return np.array(y_pred)

    def _predict(self, x):
        posteriors = []

        # Calcola la probabilità a posteriori per ogni classe
        for idx, c in enumerate(self.classes):
            prior = np.log(self.priors[idx])
            posterior = np.sum(np.log(self._pdf(idx, x)))
            posterior = prior + posterior
            posteriors.append(posterior)

        # Ritorna la classe con la probabilità a posteriori più alta
        return self.classes[np.argmax(posteriors)]

    def _pdf(self, class_idx, x):
        mean = self.mean[class_idx]
        var = self.var[class_idx]
        numerator = np.exp(-((x - mean) ** 2) / (2 * var))
        denominator = np.sqrt(2 * np.pi * var)
        return numerator / denominator
```

- #### Random Committee

    - Poiché non esiste un algoritmo standard chiamato "Random Committee" nell'ambito dell'apprendimento automatico, è stata utilizzata una tecnica di aggregazione di modelli che può essere chiamata "comitati casuali". In questo caso si è utilizzato l’algoritmo Bagging (Bootstrap Aggregating). Questo algoritmo è un metodo di apprendimento ensemble che consiste nel creare molteplici istanze di un modello di apprendimento automatico, ciascuna addestrata su un subset casuale dei dati di addestramento. In questo modo, si riduce il rischio di overfitting e si aumenta la capacità di generalizzazione del modello.
    Di seguito, un esempio di codice per l'algoritmo Bagging in Python utilizzando la libreria scikit-learn.


    - In questo esempio, viene utilizzata la classe BaggingClassifier della libreria scikit-learn per creare un'istanza del classificatore Bagging con 100 stimatori (il parametro n_estimators). Utilizziamo il classificatore Decision Tree come stimatore di base (il parametro base_estimator). Addestriamo il classificatore sui dati di addestramento X_train e le loro etichette y_train utilizzando il metodo fit. Successivamente, prediciamo le etichette per i dati di test X_test utilizzando il metodo predict, e valutiamo la precisione del classificatore utilizzando il metodo score, che restituisce la frazione di predizioni corrette.
    È importante notare che questo è solo un esempio di implementazione dell'algoritmo Bagging e che ci sono molte opzioni di configurazione e iperparametri che possono influire sulle prestazioni del modello, come il tipo di stimatore di base, la dimensione dei subset dei dati di addestramento, e così via. Tali iperparametri possono essere ottimizzati utilizzando la tecnica della ricerca degli iperparametri.


```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

# Creare un'istanza del classificatore Decision Tree
tree = DecisionTreeClassifier()

# Creare un'istanza del classificatore Bagging con 100 stimatori
bagging = BaggingClassifier(estimator=tree, n_estimators=100)
```
