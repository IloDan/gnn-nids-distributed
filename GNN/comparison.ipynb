```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

# Configurazione
protocol = 'all'
results_path = '/gpuserver/caputo/keep_calm/GNN/results'
metrics = ['accuracy', 'f1_malicious', 'precision_malicious', 'recall_malicious', 'train_time', 'test_time', 'peak_memory']

# Caricamento CSV
ddp_files = glob.glob(f"{results_path}/ddp_training_results_{protocol}_*.csv")
single_files = glob.glob(f"{results_path}/training_results_{protocol}_*.csv")
# single_files = glob.glob(f"{results_path}/non_stratified/ddp_training_results_{protocol}_*.csv")

def load_and_tag(files, mode):
    dfs = []
    for f in files:
        df = pd.read_csv(f)
        seed = f.split('_seed')[-1].replace('.csv', '')
        df['seed'] = seed
        df['mode'] = mode
        dfs.append(df)
    return pd.concat(dfs, ignore_index=True)

# Carica i dati
df_ddp = load_and_tag(ddp_files, 'ddp')
df_single = load_and_tag(single_files, 'single_gpu')
df_all = pd.concat([df_ddp, df_single], ignore_index=True)

#  Statistiche medie
print("\n Medie per categoria di attacco:")
summary = df_all.groupby(['attack_cat', 'mode'])[metrics].mean().round(4)
summary


```

```python
#to csv
summary.to_csv(f"{results_path}/summary_{protocol}.csv")
```

```python
#  T-test
print("\n Confronto statistico (t-test) per ogni attacco:")
for attack in df_all['attack_cat'].unique():
    print(f"\n Attacco: {attack.upper()}")
    df_attack = df_all[df_all['attack_cat'] == attack]
    for metric in metrics:
        vals_ddp = df_attack[df_attack['mode'] == 'ddp'][metric]
        vals_single = df_attack[df_attack['mode'] == 'single_gpu'][metric]
        if len(vals_ddp) >= 2 and len(vals_single) >= 2:
            stat, pval = ttest_ind(vals_single, vals_ddp, equal_var=False)
            print(f"{metric:22s} | p-value = {pval:.4f} {'(significativo)' if pval < 0.05 else ''}")
        else:
            print(f"{metric:22s} | Dati insufficienti per confronto")

#  Boxplot per metrica e attacco
for metric in metrics:
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df_all, x='attack_cat', y=metric, hue='mode')
    plt.title(f'Boxplot {metric} per attacco (DDP vs Single GPU)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    output_path = f"{results_path}/boxplot_{metric}_{protocol}.png"
    plt.savefig(output_path)
    print(f" Salvato grafico: {output_path}")
    plt.close()
```
